# Drinks

## Read and save data from datasets/drinks.txt

```python
FILEPATH = "/content/drive/MyDrive/lab/data/drinks.csv"
df_drink = pd.read_csv(FILEPATH)
```

```python
df_drink.columns
```

## Which continent drinks more beer on average?

```python
df_drink.groupby("continent")["total_litres_of_pure_alcohol"].mean()
result = df_drink.groupby("continent")["total_litres_of_pure_alcohol"].mean()
result[result == result.max()]
```

## For each continent print the statistics for wine consumption.

```python
df_drink.groupby("continent")["wine_servings"].describe()
```

##  Print the mean, min and max values for spirit consumption per continent.

```python
result = df_drink.groupby("continent")["spirit_servings"].agg(["mean", "min", "max"])
result
```
