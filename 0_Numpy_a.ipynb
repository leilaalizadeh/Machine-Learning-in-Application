#### Replace all odd numbers in arr with -1

```python
# Input
arr = np.arange(10)
print(arr)
# Solution
for i in arr:
  if arr[i] % 2 != 0:
    arr[i] = -1

print(arr)
```

```python
# Input
arr = np.arange(10)
print(arr)

# Solution
np.reshape(arr, (2, -1))
```

#### Stack arrays a and b vertically and then horizontally

```python
# Input
a = np.arange(10).reshape(2, -1)
b = np.repeat(1, 10).reshape(2, -1)

# Solution
hstack = np.concatenate([a, b], axis=1)
hstack1 = np.hstack([a,b])
print(hstack)
print(hstack1)
vstack = np.concatenate([a, b], axis=0)
vstack1 = np.vstack([a, b])
print(vstack)
print(vstack1)
```

#### Find positions where elements of a and b match

```python
# Input
a = np.array([1, 2, 3, 2, 3, 4, 3, 4, 5, 6])
b = np.array([7, 2, 10, 2, 7, 4, 9, 4, 9, 8])

# Solution
result = np.where(a == b)
print(result)
```

####  Get all items between 5 and 10 from a

#### Get the positions of top 5 maximum values in a given array a

```python
# Input
np.random.seed(100)
a = np.random.uniform(1, 50, 20)
print(a)

# Solution:
result = a.argsort()[-5:][::-1]
print(result)
```

#### Compute the min-by-max (i.e. min/max) for each row for given 2d numpy array (see [apply_along_axis](https://numpy.org/doc/stable/reference/generated/numpy.apply_along_axis.html#numpy-apply-along-axis))

```python
# Input
np.random.seed(100)
a = np.random.randint(1,10, [5,3])
print(a)

# Solution
max = np.amax(a, axis=1)
print(max)
min = np.amin(a, axis=1)
print(min)
result = max / min
print(result)
```

####  Find the duplicate entries in the given numpy array and mark them as True. First time occurrences should be False.

```python
# Input
np.random.seed(100)
a = np.random.randint(0, 5, 10)
print('Array: ', a)

# Solution
result = np.full(a.shape, True)
unique = np.unique(a, return_index=True)[1]
result[unique] = False
print(result)
```

#### Find the index of 5th repetition of number 1 in x
Make use of numpy.where

```python
# Input
x = np.array([1, 2, 1, 1, 3, 4, 3, 1, 1, 2, 1, 1, 2])
n = 5
# Soluzione
result = np.where(x == 1)[0]
print(result[n-1])
```

#### Compute the moving average of window size 3, for the given 1D array
Make use of numpy.convolve

```python
# Input
arr = np.array([8, 8, 3, 7, 7, 0, 4, 2, 5, 2])

# Solution
result  = np.convolve(arr, np.ones((3,))/3, mode='valid')
print(result)
```

#### Find the mean, median, standard deviation of iris's sepallength (1st column)
See np.genfromtxt, load only the 1st column with the appropiate data type

```python
# Solution
sepallength  =  np.genfromtxt(FILEPATH, delimiter=',', dtype= float, usecols=[0])


mean_val = np.mean(sepallength)
print("mean: ", mean_val)
median_val = np.median(sepallength)
print("median: ", median_val)
std_val = np.std(sepallength)
print("std: ", std_val)
```

####  Find the mean, median, standard deviation of iris's sepallength (1st column) for each unique labels

```python
# Solution
iris_data = np.genfromtxt(FILEPATH, delimiter=',', dtype=None, encoding='utf-8')

lengths = np.array([row[0] for row in iris_data])
species = np.array([row[4] for row in iris_data])


unique_species = np.unique(species)



for s in unique_species:
    mask = (species == s)

    s_lengths = lengths[mask]

    print(f"{s:<16} mean: {np.mean(s_lengths):.2f} | median: {np.median(s_lengths):.2f} | std: {np.std(s_lengths):.2f}")

```

####  Create a new column for volume in iris, where volume is (pi x petallength x sepal_length^2)/3

```python
# Solution
sepal_length = iris[:, 0].astype(float)
petal_length = iris[:, 2].astype(float)

volume = (np.pi * petal_length * (sepal_length ** 2)) / 3

iris_with_volume = np.column_stack((iris, volume))

print(iris_with_volume[:2])
```

#### Randomly sample iris's species such that setosa is twice the number of versicolor and virginica

#### Find the most frequent value of petal length (3rd column) in iris dataset

```python
# Solution
petal_length = iris[:, 2].astype(float)
unique_values, counts = np.unique(petal_length, return_counts=True)
most_frequent_value = unique_values[np.argmax(counts)]
print(most_frequent_value)
```
